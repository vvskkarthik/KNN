```python
# Set the working directory
import os
os.chdir("C:\\Users\\lenovo\\Desktop\\Praxis\\IML\\Class_Files")
#Read the cars.csv data
import pandas as pd
cars = pd.read_csv("cars.csv")
import numpy as np
import pandas as pd
```

```python
# calculating mean of the predictors and Target.

cars_mean = cars[['Horsepower','Weight','Acceleration','Displacement',]].mean()


# Calculationg standard deviation of predictors and Target.

cars_stdev = cars[['Horsepower','Weight','Acceleration','Displacement',]].std()
```

```python
# Final standardised Predictors.

cars_prdictors = {'Horsepower': (cars['Horsepower']-cars_mean[0])/cars_stdev[0],
              'Weight': (cars['Weight']-cars_mean[1])/cars_stdev[1],
              'Acceleration': (cars['Acceleration']-cars_mean[2])/cars_stdev[2], 
              'Displacement' : (cars['Acceleration']-cars_mean[3])/cars_stdev[3],
             }
cars_prdictors = pd.DataFrame(cars_prdictors)
```

```python
# Defining 
X_Predictors =cars_prdictors[['Weight','Horsepower','Displacement','Acceleration']]
Y_Target =cars['MPG']
```

```python
# Splitting into train and test data of 300 and 106 samples.

X_train=X_Predictors.iloc[:301,:]
X_test=X_Predictors.iloc[301:,:]
y_train=Y_Target.iloc[:301,]
y_test=Y_Target.iloc[301:,]
```

# A : OneNN for all predictor Test (X_Test) values

```python
# For all values in X_Test

def OneNN(X_train,X_test):
    y_pred =[]                                # Defining list to store all  predicted MPG values
    for i in range (0,len(X_test)):
        Euc_dist = []                       # Defining the list in each iteration to store the Euclidean distances
                                            # If not defined here length of this list becomes above 31,000 and returns out of bound error
        for j in range(0,len(X_train)):
            Euc_dist.append(np.sqrt(sum((X_train.iloc[j]-X_test.iloc[i])**2)))  # Calculating Euclidean Distances
            min_inedx =Euc_dist.index(min(Euc_dist))
        y_pred.append(y_train.iloc[min_inedx])   # Appending MPG values to list after each iteration.
        Final_Result = {'Predicted_MPG':y_pred,'Test_Data_MPG':y_test}
    return pd.DataFrame(Final_Result).head(10)   # Converted to dataframe for display purpose.
OneNN(X_train,X_test)
```

# B : KNN  for all predictor Test (X_Test) values

```python
def KNN(X_train,X_test,k):
    y_pred =[]                              # definig list for predicted values
    for i in range(0,len(X_test)):
        Euc_dist = []                       # Defining list to store Euclidean distances.
        Euc_dist_k ={}                      # Defining dictonary to store Euclidean distances.
        first_k_values={}                  # defining dictonary to store first k closest values of euclidean distances and their indexes
        for j in range(0,len(X_train)):
            Euc_dist.append(np.sqrt(sum((X_train.iloc[j]-X_test.iloc[i])**2)))   # Calculating Euclidean distances.
        for i in range(0,len(Euc_dist)):
            Euc_dist_k[Euc_dist.index(Euc_dist[i])] = Euc_dist[i]
        sorted_values = sorted(Euc_dist_k.values())                      # sorting the complete dictonary of euclidean diatances and their indixes
        sorted_Euc_dict ={}                                       
        for p in sorted_values:
            for l in Euc_dist_k.keys():                   
                if Euc_dist_k[l] == p:
                    sorted_Euc_dict[l] = Euc_dist_k[l]
                    break                                            # storing first k close euclidean distances and their indexes
        first_k_values = {x: sorted_Euc_dict[x] for x in list(sorted_Euc_dict)[:k]}  # list(sorted_Euc_dict) returns keys
        y_pred.append(round(y_train.iloc[list(first_k_values.keys())[0:k]].mean(),2)) # calculating prediction(MPG) by taking the mean of 'k' MPG's in y_train obtained from the indexes which are present as keys in first_k_values dictonary
    y_pred = pd.DataFrame(y_pred)
    return y_pred


min_error = 100                             # Randomly Defining Minimum Error (MSE)
for k in range(1,11):                       # Caculating all predictions(MPG) for different k values 
    y_pred =KNN(X_train,X_test,k)           # Calling the KNN function
    for w in range(0,105):
        mse = np.sqrt(sum((y_test.iloc[w]-y_pred.iloc[w])**2)/len(y_pred))   # Calculating Mean Squared Error (MSE) for different 'k' values
    if min_error > mse :
        min_error = mse
        k_best = k                          # Finding the best fit 'k' based on MSE values for different k values.  
    print("The Error for K = %d is %f" %(k,mse))
print("The Bset K value is :",k_best)
```
